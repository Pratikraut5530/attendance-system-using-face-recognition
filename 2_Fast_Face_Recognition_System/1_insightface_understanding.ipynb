```python
img = cv2.imread('test_image_2.jpg')
cv2.imshow('image',img)
cv2.waitKey(0) #to close the window
cv2.destroyAllWindows()
```

3.Apply image to insightface model and get results

```python
results_l = app_l.get(img)
```

```python
#drawing the bounding box of each detected face
img_copy = img.copy()
gender_encode = ['Female','Male']
for res in results_l:
    x1, y1, x2, y2 = res['bbox'].astype(int)
    
    #draw rectange
    cv2.rectangle(img_copy,(x1,y1),(x2,y2),(0,255,0),3)

    #draw key points
    kps = res['kps'].astype(int)
    for k1,k2 in kps:
        cv2.circle(img_copy,(k1,k2),2,(0,255,255),-1)

    #detection score 
    score = "score: {}%".format(int(res['det_score']*100))
    cv2.putText(img_copy,score,(x1,y1),cv2.FONT_HERSHEY_PLAIN,1,(0,0,0))

    ##age and gender
    gender = gender_encode[res['gender']]
    age = res['age']
    age_gender = f"{gender} :: {age}"
    cv2.putText(img_copy,age_gender,(x2,y2),cv2.FONT_HERSHEY_SIMPLEX,0.4,(255,200,0))


cv2.imshow('bbox',img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
